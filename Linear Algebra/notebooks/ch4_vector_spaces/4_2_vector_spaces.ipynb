### What you will learn

- Key **properties** that follow from the axioms: $0 \cdot \mathbf{v} = \mathbf{0}$, $c \cdot \mathbf{0} = \mathbf{0}$, $(-1) \cdot \mathbf{v} = -\mathbf{v}$.

---

### Motivation

> $\mathbb{R}^n$, matrices, polynomials, continuous functions all satisfy 10 axioms.  
> What structure do they share?

The answer is a **vector space** — an abstract algebraic structure defined by a set $V$,
two operations (addition and scalar multiplication), and 10 axioms. Once we know a set
forms a vector space, every theorem about vector spaces applies to it automatically.
This abstraction is the engine that makes linear algebra so broadly useful.

```python
from linalg_core.matrix import Matrix
from linalg_core.ops import add, scalar_mul, multiply, transpose
from linalg_core.determinant import det_elimination
from linalg_core import EPSILON

import random
random.seed(42)

def approx_eq(a, b, tol=EPSILON):
    """Check approximate equality for floats."""
    return abs(a - b) < tol

def mat_approx_eq(A, B, tol=EPSILON):
    """Check approximate equality for two matrices."""
    if A.rows != B.rows or A.cols != B.cols:
        return False
    return all(abs(a - b) < tol for a, b in zip(A.data, B.data))

def vec_approx_eq(u, v, tol=EPSILON):
    """Check approximate equality for two lists (vectors)."""
    return len(u) == len(v) and all(abs(a - b) < tol for a, b in zip(u, v))

print("Setup complete.")
```

## 2. The 10 Vector Space Axioms

A **vector space** over the real numbers is a set $V$ equipped with two operations:

- **Vector addition:** $\mathbf{u} + \mathbf{v} \in V$
- **Scalar multiplication:** $c\,\mathbf{u} \in V$

satisfying the following **10 axioms** for all $\mathbf{u}, \mathbf{v}, \mathbf{w} \in V$ and all scalars $c, d \in \mathbb{R}$:

---

**Addition axioms:**

| # | Name | Axiom |
|:-:|:-----|:------|
| 1 | Closure under addition | $\mathbf{u} + \mathbf{v} \in V$ |
| 2 | Commutativity | $\mathbf{u} + \mathbf{v} = \mathbf{v} + \mathbf{u}$ |
| 3 | Associativity of addition | $(\mathbf{u} + \mathbf{v}) + \mathbf{w} = \mathbf{u} + (\mathbf{v} + \mathbf{w})$ |
| 4 | Additive identity | There exists $\mathbf{0} \in V$ such that $\mathbf{u} + \mathbf{0} = \mathbf{u}$ |
| 5 | Additive inverse | For each $\mathbf{u}$, there exists $-\mathbf{u} \in V$ such that $\mathbf{u} + (-\mathbf{u}) = \mathbf{0}$ |

**Scalar multiplication axioms:**

| # | Name | Axiom |
|:-:|:-----|:------|
| 6 | Closure under scalar multiplication | $c\,\mathbf{u} \in V$ |
| 7 | Distributivity over vector addition | $c(\mathbf{u} + \mathbf{v}) = c\,\mathbf{u} + c\,\mathbf{v}$ |
| 8 | Distributivity over scalar addition | $(c + d)\,\mathbf{u} = c\,\mathbf{u} + d\,\mathbf{u}$ |
| 9 | Associativity of scalar multiplication | $c(d\,\mathbf{u}) = (cd)\,\mathbf{u}$ |
| 10 | Multiplicative identity | $1 \cdot \mathbf{u} = \mathbf{u}$ |

---

> **Key insight:** These axioms are *not* properties to memorize — they are the *definition*.
> A set either satisfies all 10, earning the title "vector space," or it doesn't.
> One failure is enough to disqualify.

## 3. Verify $\mathbb{R}^3$ — The Prototypical Vector Space

We represent vectors in $\mathbb{R}^3$ as Python lists `[x, y, z]` and verify all
10 axioms computationally with random vectors.

```python
def rand_vec3():
    """Generate a random vector in R^3 with entries in [-10, 10]."""
    return [random.uniform(-10, 10) for _ in range(3)]

def vec_add(u, v):
    """Component-wise addition of two R^n vectors."""
    return [a + b for a, b in zip(u, v)]

def vec_smul(c, u):
    """Scalar multiplication for R^n vectors."""
    return [c * a for a in u]

def vec_neg(u):
    """Additive inverse."""
    return [-a for a in u]

zero3 = [0.0, 0.0, 0.0]

u = rand_vec3()
v = rand_vec3()
w = rand_vec3()
c = random.uniform(-5, 5)
d = random.uniform(-5, 5)

print(f"u = {[round(x,4) for x in u]}")
print(f"v = {[round(x,4) for x in v]}")
print(f"w = {[round(x,4) for x in w]}")
print(f"c = {c:.4f}")
print(f"d = {d:.4f}")
print()

results = {}

# Axiom 1: Closure under addition (u + v is a list of length 3)
s = vec_add(u, v)
results["A1 Closure (+)"] = len(s) == 3 and all(isinstance(x, float) for x in s)

# Axiom 2: Commutativity
results["A2 Commutativity"] = vec_approx_eq(vec_add(u, v), vec_add(v, u))

# Axiom 3: Associativity of addition
results["A3 Assoc (+)"] = vec_approx_eq(
    vec_add(vec_add(u, v), w),
    vec_add(u, vec_add(v, w))
)

# Axiom 4: Additive identity
results["A4 Zero vec"] = vec_approx_eq(vec_add(u, zero3), u)

# Axiom 5: Additive inverse
results["A5 Inverse"] = vec_approx_eq(vec_add(u, vec_neg(u)), zero3)

# Axiom 6: Closure under scalar multiplication
cu = vec_smul(c, u)
results["A6 Closure (*)"] = len(cu) == 3 and all(isinstance(x, float) for x in cu)

# Axiom 7: Distributivity over vector addition
results["A7 c(u+v)"] = vec_approx_eq(
    vec_smul(c, vec_add(u, v)),
    vec_add(vec_smul(c, u), vec_smul(c, v))
)

# Axiom 8: Distributivity over scalar addition
results["A8 (c+d)u"] = vec_approx_eq(
    vec_smul(c + d, u),
    vec_add(vec_smul(c, u), vec_smul(d, u))
)

# Axiom 9: Associativity of scalar multiplication
results["A9 c(du)"] = vec_approx_eq(
    vec_smul(c, vec_smul(d, u)),
    vec_smul(c * d, u)
)

# Axiom 10: Multiplicative identity
results["A10 1*u=u"] = vec_approx_eq(vec_smul(1.0, u), u)

print("=== R^3 Axiom Verification ===")
all_pass = True
for name, passed in results.items():
    status = "PASS" if passed else "FAIL"
    if not passed:
        all_pass = False
    print(f"  {name:25s} {status}")

print()
print("Conclusion:", "R^3 is a vector space." if all_pass else "R^3 FAILED an axiom!")
```

## 4. Verify $M_{2 \times 2}$ — The Space of $2 \times 2$ Matrices

The set of all $2 \times 2$ real matrices forms a vector space under the usual
matrix addition and scalar multiplication.

We use the `Matrix` class and `ops.add` / `ops.scalar_mul` from `linalg_core`.

```python
def rand_mat22():
    """Generate a random 2x2 matrix with entries in [-10, 10]."""
    return Matrix.from_lists([
        [random.uniform(-10, 10), random.uniform(-10, 10)],
        [random.uniform(-10, 10), random.uniform(-10, 10)]
    ])

zero22 = Matrix(2, 2)  # 2x2 zero matrix

A = rand_mat22()
B = rand_mat22()
C = rand_mat22()
c = random.uniform(-5, 5)
d = random.uniform(-5, 5)

print("A =", A)
print("B =", B)
print("C =", C)
print(f"c = {c:.4f}, d = {d:.4f}")
print()

results = {}

# Axiom 1: Closure — add returns a 2x2 Matrix
S = add(A, B)
results["A1 Closure (+)"] = S.rows == 2 and S.cols == 2

# Axiom 2: Commutativity
results["A2 Commutativity"] = mat_approx_eq(add(A, B), add(B, A))

# Axiom 3: Associativity of addition
results["A3 Assoc (+)"] = mat_approx_eq(add(add(A, B), C), add(A, add(B, C)))

# Axiom 4: Additive identity
results["A4 Zero mat"] = mat_approx_eq(add(A, zero22), A)

# Axiom 5: Additive inverse  (-1 * A is the additive inverse)
neg_A = scalar_mul(A, -1.0)
results["A5 Inverse"] = mat_approx_eq(add(A, neg_A), zero22)

# Axiom 6: Closure under scalar multiplication
cA = scalar_mul(A, c)
results["A6 Closure (*)"] = cA.rows == 2 and cA.cols == 2

# Axiom 7: c(A + B) = cA + cB
results["A7 c(A+B)"] = mat_approx_eq(
    scalar_mul(add(A, B), c),
    add(scalar_mul(A, c), scalar_mul(B, c))
)

# Axiom 8: (c + d)A = cA + dA
results["A8 (c+d)A"] = mat_approx_eq(
    scalar_mul(A, c + d),
    add(scalar_mul(A, c), scalar_mul(A, d))
)

# Axiom 9: c(dA) = (cd)A
results["A9 c(dA)"] = mat_approx_eq(
    scalar_mul(scalar_mul(A, d), c),
    scalar_mul(A, c * d)
)

# Axiom 10: 1 * A = A
results["A10 1*A=A"] = mat_approx_eq(scalar_mul(A, 1.0), A)

print("=== M_{2x2} Axiom Verification ===")
all_pass = True
for name, passed in results.items():
    status = "PASS" if passed else "FAIL"
    if not passed:
        all_pass = False
    print(f"  {name:25s} {status}")

print()
print("Conclusion:", "M_{2x2} is a vector space." if all_pass else "M_{2x2} FAILED an axiom!")
```

## 5. Verify $P_2$ — Polynomials of Degree $\leq 2$

A polynomial $p(x) = a + bx + cx^2$ is represented as the coefficient list `[a, b, c]`.

Addition and scalar multiplication are defined component-wise on the coefficients,
which mirrors the algebraic rules:

$$
(a_1 + b_1 x + c_1 x^2) + (a_2 + b_2 x + c_2 x^2) = (a_1 + a_2) + (b_1 + b_2)x + (c_1 + c_2)x^2
$$

```python
def rand_poly2():
    """Random polynomial in P_2: [a, b, c] represents a + bx + cx^2."""
    return [random.uniform(-10, 10) for _ in range(3)]

def poly_add(p, q):
    """Add two polynomials in P_2."""
    return [a + b for a, b in zip(p, q)]

def poly_smul(c, p):
    """Scalar multiply a polynomial in P_2."""
    return [c * a for a in p]

def poly_neg(p):
    """Additive inverse of a polynomial."""
    return [-a for a in p]

def poly_str(p):
    """Pretty-print a polynomial."""
    a, b, c = [round(x, 4) for x in p]
    return f"{a} + {b}x + {c}x^2"

zero_p = [0.0, 0.0, 0.0]

p = rand_poly2()
q = rand_poly2()
r = rand_poly2()
c = random.uniform(-5, 5)
d = random.uniform(-5, 5)

print(f"p(x) = {poly_str(p)}")
print(f"q(x) = {poly_str(q)}")
print(f"r(x) = {poly_str(r)}")
print(f"c = {c:.4f}, d = {d:.4f}")
print()

results = {}

# Axiom 1: Closure under addition (result has 3 coefficients)
s = poly_add(p, q)
results["A1 Closure (+)"] = len(s) == 3

# Axiom 2: Commutativity
results["A2 Commutativity"] = vec_approx_eq(poly_add(p, q), poly_add(q, p))

# Axiom 3: Associativity of addition
results["A3 Assoc (+)"] = vec_approx_eq(
    poly_add(poly_add(p, q), r),
    poly_add(p, poly_add(q, r))
)

# Axiom 4: Additive identity (0 + 0x + 0x^2)
results["A4 Zero poly"] = vec_approx_eq(poly_add(p, zero_p), p)

# Axiom 5: Additive inverse
results["A5 Inverse"] = vec_approx_eq(poly_add(p, poly_neg(p)), zero_p)

# Axiom 6: Closure under scalar multiplication
cp = poly_smul(c, p)
results["A6 Closure (*)"] = len(cp) == 3

# Axiom 7: c(p + q) = cp + cq
results["A7 c(p+q)"] = vec_approx_eq(
    poly_smul(c, poly_add(p, q)),
    poly_add(poly_smul(c, p), poly_smul(c, q))
)

# Axiom 8: (c + d)p = cp + dp
results["A8 (c+d)p"] = vec_approx_eq(
    poly_smul(c + d, p),
    poly_add(poly_smul(c, p), poly_smul(d, p))
)

# Axiom 9: c(dp) = (cd)p
results["A9 c(dp)"] = vec_approx_eq(
    poly_smul(c, poly_smul(d, p)),
    poly_smul(c * d, p)
)

# Axiom 10: 1 * p = p
results["A10 1*p=p"] = vec_approx_eq(poly_smul(1.0, p), p)

print("=== P_2 Axiom Verification ===")
all_pass = True
for name, passed in results.items():
    status = "PASS" if passed else "FAIL"
    if not passed:
        all_pass = False
    print(f"  {name:25s} {status}")

print()
print("Conclusion:", "P_2 is a vector space." if all_pass else "P_2 FAILED an axiom!")
```

## 6. Non-Example: $2 \times 2$ Matrices with $\det = 1$

Let $S = \{A \in M_{2 \times 2} : \det(A) = 1\}$ with the usual matrix addition.

**Claim:** $S$ is **not** a vector space because it fails **closure under addition**.

If $\det(A) = 1$ and $\det(B) = 1$, it is generally **not** true that $\det(A + B) = 1$.

```python
A = Matrix.from_lists([[1, 0],
                        [0, 1]])   # Identity matrix, det = 1

B = Matrix.from_lists([[0, 1],
                        [-1, 0]])  # Rotation by 90 degrees, det = 1

det_A = det_elimination(A)
det_B = det_elimination(B)

ApB = add(A, B)
det_ApB = det_elimination(ApB)

print("A =", A)
print(f"det(A) = {det_A:.4f}")
print()
print("B =", B)
print(f"det(B) = {det_B:.4f}")
print()
print("A + B =", ApB)
print(f"det(A + B) = {det_ApB:.4f}")
print()

if approx_eq(det_A, 1.0) and approx_eq(det_B, 1.0):
    print("Both A and B have det = 1, so both are in S.")
else:
    print("ERROR: A or B does not have det = 1.")

if not approx_eq(det_ApB, 1.0):
    print(f"But det(A + B) = {det_ApB:.4f} != 1.")
    print("CLOSURE UNDER ADDITION FAILS.")
    print()
    print("Conclusion: S = {A in M_{2x2} : det(A) = 1} is NOT a vector space.")
else:
    print("det(A + B) = 1 for this example (try different matrices).")
```

> **Note:** The set $S$ also fails to contain the zero matrix ($\det(\mathbf{0}) = 0 \neq 1$),
> so Axiom 4 (additive identity) fails as well. Any single axiom failure is sufficient
> to disqualify $S$ as a vector space.

## 7. Properties from the Axioms

The following properties can be **proved** from the 10 axioms alone. They hold
in *every* vector space.

| Property | Statement |
|:---------|:----------|
| **P1** | $0 \cdot \mathbf{v} = \mathbf{0}$ (the scalar zero times any vector is the zero vector) |
| **P2** | $c \cdot \mathbf{0} = \mathbf{0}$ (any scalar times the zero vector is the zero vector) |
| **P3** | $(-1) \cdot \mathbf{v} = -\mathbf{v}$ (scalar $-1$ gives the additive inverse) |

We verify these computationally in all three spaces.

```python
print("=" * 55)
print("Property Verification Across Three Vector Spaces")
print("=" * 55)

# --- R^3 ---
v3 = rand_vec3()
print(f"\n--- R^3 with v = {[round(x,4) for x in v3]} ---")

# P1: 0 * v = 0
p1_r3 = vec_approx_eq(vec_smul(0.0, v3), zero3)
print(f"  P1  0*v = 0          : {vec_smul(0.0, v3)} -> {'PASS' if p1_r3 else 'FAIL'}")

# P2: c * 0 = 0
c_test = random.uniform(-5, 5)
p2_r3 = vec_approx_eq(vec_smul(c_test, zero3), zero3)
print(f"  P2  c*0 = 0  (c={c_test:.2f}) : {vec_smul(c_test, zero3)} -> {'PASS' if p2_r3 else 'FAIL'}")

# P3: (-1)*v = -v
p3_r3 = vec_approx_eq(vec_smul(-1.0, v3), vec_neg(v3))
print(f"  P3  (-1)*v = -v      : {'PASS' if p3_r3 else 'FAIL'}")

# --- M_{2x2} ---
M = rand_mat22()
print(f"\n--- M_{{2x2}} with M = ")
print(M)

# P1: 0 * M = zero matrix
p1_m = mat_approx_eq(scalar_mul(M, 0.0), zero22)
print(f"  P1  0*M = 0          : {'PASS' if p1_m else 'FAIL'}")

# P2: c * 0 = zero matrix
p2_m = mat_approx_eq(scalar_mul(zero22, c_test), zero22)
print(f"  P2  c*0 = 0  (c={c_test:.2f}) : {'PASS' if p2_m else 'FAIL'}")

# P3: (-1)*M = -M (additive inverse)
neg_M = scalar_mul(M, -1.0)
p3_m = mat_approx_eq(add(M, neg_M), zero22)
print(f"  P3  (-1)*M = -M      : {'PASS' if p3_m else 'FAIL'}")

# --- P_2 ---
pp = rand_poly2()
print(f"\n--- P_2 with p(x) = {poly_str(pp)} ---")

# P1: 0 * p = zero polynomial
p1_p = vec_approx_eq(poly_smul(0.0, pp), zero_p)
print(f"  P1  0*p = 0          : {'PASS' if p1_p else 'FAIL'}")

# P2: c * 0 = zero polynomial
p2_p = vec_approx_eq(poly_smul(c_test, zero_p), zero_p)
print(f"  P2  c*0 = 0  (c={c_test:.2f}) : {'PASS' if p2_p else 'FAIL'}")

# P3: (-1)*p = -p
p3_p = vec_approx_eq(poly_smul(-1.0, pp), poly_neg(pp))
print(f"  P3  (-1)*p = -p      : {'PASS' if p3_p else 'FAIL'}")

print("\n" + "=" * 55)
all_props = all([p1_r3, p2_r3, p3_r3, p1_m, p2_m, p3_m, p1_p, p2_p, p3_p])
print("All properties verified:", "YES" if all_props else "NO")
```
